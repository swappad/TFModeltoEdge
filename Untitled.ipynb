```python
KERNEL_NAME = "Unet"
KERNEL_CONV_INPUT = "block1_conv1_Conv2D"
KERNEL_OUTPUT = "conv2d_8_Conv2D"
```

```python
n2cube.dpuOpen()
kernel = n2cube.dpuLoadKernel(KERNEL_NAME)
task = n2cube.dpuCreateTask(kernel, 0)
```

```python
input_len = n2cube.dpuGetInputTensorSize(task, KERNEL_CONV_INPUT)
output_len = n2cube.dpuGetOutputTensorSize(task, KERNEL_OUTPUT)
conf = n2cube.dpuGetOutputTensorAddress(task, KERNEL_OUTPUT)
output_scale = n2cube.dpuGetOutputTensorScale(task, KERNEL_OUTPUT)
print(input_len, output_len, conf, output_scale)
```

```python
import numpy as np
import os
from time import time
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
def result_map_to_img(res_map):
    img = np.zeros((256, 512, 3), dtype=np.uint8)
    res = np.squeeze(res_map)
  
    argmax_idx = np.argmax(res, axis=2)
  
    # For np.where calculation.
    person = (argmax_idx == 1)
    car = (argmax_idx == 2)
    road = (argmax_idx == 3)
  
    img[:, :, 0] = np.where(person, 255, 0)
    img[:, :, 1] = np.where(car, 255, 0)
    img[:, :, 2] = np.where(road, 255, 0)
  
    return img
```

```python
image = cv2.imread("/home/xilinx/jupyter_notebooks/own/leftImg8bit/train/aachen/aachen_000008_000019_leftImg8bit.png")

try:
    image_small = cv2.resize(image, (512, 256))
    #image_small = cv2.cvtColor(image_small, cv2.COLOR_BGR2RGB)
    #image_small = image_small.transpose((2,0,1)) 
    #image_small = image_small / 127.5 - 1
    image_small = np.expand_dims(image_small, 0)
    image_data = np.array(image_small, dtype=np.int8)
    plt.imshow(image)
except cv2.error as e:
    print('Invald frame!')
    print(e)
image_data.shape
```

```python
calib_img_path = "./leftImg8bit/"
calib_batch_size = 1
## images for evaluation
paths = []
for (path, dirname, files) in sorted(os.walk(calib_img_path)):
    for filename in sorted(files):
        if filename.endswith(('.jpg', '.png')):
            paths.append(os.path.join(path, filename))
images = []
for index in range(0, calib_batch_size):
    #print("Path: " + paths[(iter * calib_batch_size + index) % len(paths)])
    img = cv2.imread(paths[(100 * calib_batch_size + index) % len(paths)], 1)
    try:
        img = cv2.resize(img,(512, 256))
        #img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        #img = img / 127.5 - 1
        images.append(img)
        plt.imshow(img)
    except cv2.error as e:
            print('Invalid frame!')
            print(e)
image_data = np.expand_dims(images[0], 0)
image_data.dtype
```

```python
n2cube.dpuSetInputTensorInCHWInt8(task, KERNEL_CONV_INPUT, image_data.astype(np.int8), input_len)
n2cube.dpuRunTask(task)
image_out = n2cube.dpuGetOutputTensorInHWCInt8(task, KERNEL_OUTPUT, output_len)
```

```python
classes = np.array(image_out, dtype=np.float32)
classes = np.reshape(classes, (256, 512, 4))
result = result_map_to_img(classes.astype(np.uint8))
plt.imshow(result)
```
